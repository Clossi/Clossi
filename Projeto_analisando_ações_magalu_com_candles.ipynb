```python
# Libs para Modelagem e Matriz

import numpy as np
import pandas as pd


# Libs para Analises Graficas

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

# Libs para ignorar avisos

import warnings

# Desabilitar avisos
warnings.filterwarnings('ignore')
```

```python
# Lendo os dados
Base_Dados = pd.read_excel('Vase_004 - Magalu - Sem Resolução.xlsx')

# Verificando
Base_Dados.head()
```

```python
Base_Dados.shape
```

```python
Base_Dados.info()
```

```python
Base_Dados.describe()
```

```python
# Series temporais
Dados = Base_Dados.set_index('Data')

Dados.head()
```

```python
plt.style.use('seaborn-darkgrid')
plt.figure( figsize=(16, 5) )
plt.title('Analise das ações da Magalu - Fechamento', fontsize=16, loc='left')
plt.plot( Dados.index, Dados['Fechamento'])
plt.xlabel('Periodo da Cotação', fontsize=14, )
plt.ylabel('Valor da ação (R$)', fontsize=14 );
```

```python
Dados.tail()
```

```python
Media_Movel = Dados['Fechamento'].rolling(5).mean()
Media_Tendencia = Dados['Fechamento'].rolling(30).mean()

plt.style.use('seaborn-darkgrid')
plt.figure( figsize=(16, 5) )
plt.title('Analise das ações da Magalu - Fechamento', fontsize=16, loc='left')

plt.plot( Dados.index, Dados['Fechamento'])
plt.plot( Dados.index, Media_Movel )
plt.plot( Dados.index, Media_Tendencia )


plt.xlabel('Periodo da Cotação', fontsize=14, )
plt.ylabel('Valor da ação (R$)', fontsize=14 );
```

```python
sns.boxplot( data=Dados, x='Fechamento');
```

```python
# BoxPlot Mensal

Base_Dados['Mes'] = Base_Dados['Data'].dt.month

plt.figure( figsize=(16, 5) )
sns.boxplot( data=Base_Dados, x='Mes', y='Fechamento' );
```

```python
Base_Dados.groupby(['Mes']).describe()['Fechamento']
```

```python
# Grafico de Candles/Velas
Grafico = go.Figure(
    data=[
        go.Candlestick(
            x= Dados.index,
            open = Dados['Abertura'],
            high = Dados['Maior'],
            low = Dados['Menor'],
            close = Dados['Fechamento'],
        )
    ]
)

Grafico.update_layout( xaxis_rangeslider_visible=False )

Grafico.show()
```
