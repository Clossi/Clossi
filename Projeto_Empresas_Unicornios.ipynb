```python
# Primeiro eu importei as libs que serão usadas neste projeto

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Este abaixo é só para não ficar dando avisos chatos
import warnings
warnings.filterwarnings('ignore')
```

```python
# Agora leitura dos dados
Base_Dados = pd.read_csv('Startups+in+2021+end.csv')

# Vou verificar a dimensão dele
Base_Dados.shape
```

```python
# A base de dados possui 936 linhas e 8 colunas
# Vamos verificar primeiros registros
Base_Dados.head()
```

```python
# Vou veriicar as colunas existentes e depois vamos renomear todas elas para português BR
Base_Dados.columns
```

```python
Base_Dados.rename(columns={
    'Unnamed: 0' : 'Id',
    'Company' : 'Empresa',
    'Valuation ($B)' : 'Valor',
    'Date Joined' : 'Data de Adesão',
    'Country'	: 'País',
    'City' : 'Cidade',
    'Industry'	: 'Setor',
    'Select Investors' : 'Investidores',
}, inplace=True )
```

```python
# Vamos gerar a base de dados para saber como ficou a tradução do cabeçalho das colunas

Base_Dados.info()
```

```python
# Vamos verificar campos nulos
Base_Dados.isnull().sum()
```

```python
#Campos nulos em grafico para visualizar

plt.figure( figsize=(13,6) )
plt.title('Analisando Campos Nulos')
sns.heatmap( Base_Dados.isnull(), cbar=False );
```

```python
# Verifcando quantos campos unicos na base de dados

Base_Dados.nunique()
```

```python
# Vou focar em verificar o Setor agora, pois é interessante saber quais setores que fazem estes volumes

Base_Dados['Setor'].value_counts( normalize=True )

# Setor separado por quantia em percentual, as Fintech somam 20% do total etc...
```

```python
# Analise grafica dos setores por quantia numerica
plt.figure( figsize=(13,6) )
plt.title('Analisando por Setores')
plt.bar( Base_Dados['Setor'].value_counts().index, Base_Dados['Setor'].value_counts() )
plt.xticks( rotation=45, ha='right' );
```

```python
# Criei uma variavel chamada Analise, pois vamos verificar agora os países
Analise = round(Base_Dados['País'].value_counts( normalize=True ) * 100, 1 )
```

```python
# Testando a variavel
Analise
```

```python
# Criando grafico da Analise em Países TOP 10 em percentual

plt.figure( figsize=(16,10) )
plt.title('Analisando TOP 10 Países em %')
plt.pie(
    Analise.head(10),
    labels = Analise.index[0:10],
    shadow=True,
    startangle=90,
    autopct='%1.1f%%'
),
plt.xticks( rotation=45, ha='left' );
```

```python
# Verifiquei que a coluna data e Valor estão em formato objeto, porque são numeros então terei que formatar
#Neste codigo formatei a Data para formato datetime
Base_Dados['Data de Adesão'] = pd.to_datetime( Base_Dados['Data de Adesão'] )
Base_Dados['Data de Adesão'].head()
```

```python
# Extrair Mês e Ano

Base_Dados['Mês'] = pd.DatetimeIndex( Base_Dados['Data de Adesão'] ).month
Base_Dados['Ano'] = pd.DatetimeIndex( Base_Dados['Data de Adesão'] ).year
Base_Dados.head()
```

```python
# Tabela Analitica na coluna Id para contagem

Analise_Agrupada = Base_Dados.groupby(by=['País', 'Ano', 'Mês', 'Empresa'] ).count()['Id'].reset_index()

Analise_Agrupada
```

```python
# Pesquisando apenas 1 país

Analise_Agrupada.loc[
    Analise_Agrupada['País'] == 'Brazil'
]
```

```python
Base_Dados['Valor'] = pd.to_numeric( Base_Dados['Valor'].apply( lambda Linha: Linha.replace('$', '') ) )

Base_Dados.head()
```

```python
Analise_País = Base_Dados.groupby( by=['País'] ).sum()['Valor'].reset_index().sort_values('Valor')

Analise_País.head()
```

```python
Analise_Valor = Analise_País.sort_values('Valor', ascending=False)

Analise_Valor.head()
```

```python

plt.figure( figsize=(13,6) )
plt.title('Analise Valor po País')
plt.xticks( rotation=45, ha='right' )
plt.plot( Analise_Valor['País'], Analise_Valor['Valor'] );
```
